```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
xmin, xmax = -1.5, 1.5
ymin, ymax = -1.5, 1.5

xwidth = xmax-xmin
yht = ymax-ymin
```

```python
n_iters_max=2000
im_width= 2000
im_ht = 1000
```

```python
julia = np.zeros((im_width, im_ht))
c = -0.1+0.651j
```

```python
for ix in range(im_width):
    for iy in range(im_ht):
        num_iters = 0
        z = complex(ix/im_width*xwidth + xmin, iy/im_ht*yht + ymin)
        while abs(z) <= 10 and num_iters < n_iters_max:
            z = z**2 + c
            num_iters+=1
        julia[ix, iy] = num_iters/n_iters_max
```

```python
plt.figure(dpi=800)
fig, ax = plt.subplots()
plt.axis('off')
ax.imshow(julia, interpolation='nearest', cmap='gnuplot', aspect='auto');
#plt.savefig('C:/Users/abaks/desktop/frac1.png', bbox_inches='tight')
```
