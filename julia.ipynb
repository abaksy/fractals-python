# Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Setting thresholds for the plot width

The values of $x_{min}$, $x_{max}$, $y_{min}$ and $y_{max}$ decide the size of the final plot

```python
xmin, xmax = -1.5, 1.5
ymin, ymax = -1.5, 1.5

xwidth = xmax-xmin
yht = ymax-ymin
```

## Parameter declaration

Setting maximum number of iterations for a particular pixel, and image dimensions

```python
n_iters_max=2000
im_width= 2000
im_ht = 1000
```

```python
c = -0.1+0.651j                     # Constant c to be used in the iterative process
```

# Populating Julia matrix

For each point $(x, y)$ in the matrix, calculate the number of iterations needed for the complex number representing that point to <br>
reach an absolute value of 10 under the iterative process $z = z^2 + c$ where c is declared as above

```python
def julia_plot(c, num_iters_max, im_width, im_height, xwidth, yht):
    julia = np.zeros((im_width, im_height)) # Initializing matrix which is to be plotted as image
    for ix in range(im_width):
        for iy in range(im_height):
            num_iters = 0
            z = complex(ix/im_width*xwidth + xmin, iy/im_height*yht + ymin)
            while abs(z) <= 10 and num_iters < n_iters_max:
                z = z**2 + c
                num_iters+=1
            julia[ix, iy] = num_iters/n_iters_max
    return julia
```

## Plot the matrix

```python
julia = julia_plot(c, n_iters_max, im_width, im_ht, xwidth, yht)
plt.figure(dpi=800)
fig, ax = plt.subplots();
plt.axis('off')
ax.imshow(julia, interpolation='nearest', cmap='gnuplot', aspect='auto');
```

## Changing the parameter $c$

```python
julia = julia_plot(complex(-0.2, 0.65), n_iters_max, im_width, im_ht, xwidth, yht)
plt.figure(dpi=800);
fig, ax = plt.subplots()
plt.axis('off')
ax.imshow(julia, interpolation='nearest', cmap='afmhot', aspect='auto');
```
